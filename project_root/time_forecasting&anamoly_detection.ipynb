## Anomaly Detection with Prophet

#### Function to Aggregate Daily Connections for a Specific l_ipn

```python
def get_daily_aggregate_l_ipn(in_l_ipn):
    temp_df = daily_aggregate_l_ipn[daily_aggregate_l_ipn['l_ipn'] == in_l_ipn].drop(['l_ipn'], axis=1)
    temp_df.columns = ['ds', 'y']
    temp_df['ds'] = to_datetime(temp_df['ds'])
    temp_df.reset_index(inplace=True, drop=True)
    return temp_df
```

#### Function to Perform Forecasting and Anomaly Detection



```python
def get_forecast(ts, in_l_ipn):
    model = Prophet(seasonality_mode='additive', daily_seasonality=False, yearly_seasonality=False, weekly_seasonality=True)
    model.fit(ts)
    forecast = model.predict(pd.DataFrame(ts['ds']))
    ts['anomaly'] = 0
    p_color = np.full((ts.shape[0], 1), 'green')
    for i in range(forecast.shape[0]):
        if (forecast.at[i, 'yhat_lower'] > ts.at[i, 'y']) or (forecast.at[i, 'yhat_upper'] < ts.at[i, 'y']):
            ts.at[i, 'anomaly'] = 1
            p_color[i] = 'red'
    model.plot(forecast)
    plt.scatter(ts['ds'], ts['y'], c=p_color.ravel())
    plt.title(f'Forecast plot for l_ipn {in_l_ipn}')
    plt.show()
```

#### Applying the Forecasting Function to Each l_ipn

```python
for i in range(df['l_ipn'].nunique()):
    get_forecast(get_daily_aggregate_l_ipn(i), i)
```

## Histograms
#### Analyzing Internet Traffic Patterns

##### The provided code aims to provide insights into internet Protocal (IP) traffic patterns by  visualizing the distribution of connection counts and ISP counts for different hosts.This code appears to be analyzing a dataset related to internet traffic, specifically focusing on the number of connections and Internet Service Provider (ISP) counts for different hosts over time.



```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random
import matplotlib.patches as patches
%matplotlib inline
# df = pd.read_csv('../input/cs448b_ipasn.csv')
df = pd.read_csv('network_data.csv')
```

```python

grouped = df[['date','l_ipn','f']].groupby(['date','l_ipn'])
res = grouped[['f']].agg(['sum','count'])['f'][['sum','count']];
res=res.rename(columns={"sum": "connections_count", "count": "isp_count"})
res = res.reset_index()  # reset grouping
df0 = res.rename(columns={"connections_count":"f", "isp_count":"cnt"})

plt.rcParams['figure.figsize'] = 15, 50
f, axes = plt.subplots(10, 2, sharey=False)

# Custom side headings for each subplot column
side_headings = ['Custom Heading 1', 'Custom Heading 2']

for i in range(10):
    b = 20
    mu_f = df0[df0.l_ipn == i]["f"].values.mean()
    s_f = df0[df0.l_ipn == i]["f"].values.std()
    mu_cnt = df0[df0.l_ipn == i]["cnt"].values.mean()
    s_cnt = df0[df0.l_ipn == i]["cnt"].values.std() 

    axes[i][0].set_title('Connections count for host: '+str(i)+', $ \mu=' + str(mu_f) + ', \sigma=' + str(s_f) +'$')
    axes[i][0].hist(df0[df0.l_ipn==i]["f"].values, bins=b, label='ISP count for host:')
    axes[i][0].axvline(mu_f, color='g', linewidth=2)
    rect = patches.Rectangle((mu_f,0), s_f, axes[i][0].get_ylim()[1], linewidth=0, facecolor='g', alpha=0.1)
    axes[i][0].add_patch(rect)
    rect = patches.Rectangle((mu_f + s_f,0), s_f, axes[i][0].get_ylim()[1], linewidth=0, facecolor='g', alpha=0.06)
    axes[i][0].add_patch(rect)
    rect = patches.Rectangle((mu_f + s_f + s_f,0), s_f, axes[i][0].get_ylim()[1], linewidth=0, facecolor='g', alpha=0.03)
    axes[i][0].add_patch(rect)

    axes[i][1].set_title('ISP count for host: '+str(i)+', $ \mu=' + str(mu_cnt) + ', \sigma=' + str(s_cnt) +'$')
    axes[i][1].hist(df0[df0.l_ipn==i]["cnt"].values, bins=b)
    axes[i][1].axvline(mu_cnt, color='g', linewidth=2)
    rect = patches.Rectangle((mu_cnt,0), s_cnt, axes[i][1].get_ylim()[1], linewidth=0, facecolor='g', alpha=0.1)
    axes[i][1].add_patch(rect)
    rect = patches.Rectangle((mu_cnt + s_cnt,0), s_cnt, axes[i][1].get_ylim()[1], linewidth=0, facecolor='g', alpha=0.06)
    axes[i][1].add_patch(rect)
    rect = patches.Rectangle((mu_cnt + s_cnt + s_cnt,0), s_cnt, axes[i][1].get_ylim()[1], linewidth=0, facecolor='g', alpha=0.03)
    axes[i][1].add_patch(rect)

# Side Headings
for ax, col in zip(axes[0], side_headings):
    ax.annotate(col, xy=(0.5, 1), xytext=(0, 20), xycoords='axes fraction', textcoords='offset points',
                size='large', ha='center', va='baseline')

plt.show()
```

## Anomaly Detection with Luminol

#### Function to Detect Anomalies using Luminol

```python
## Anomaly Detection with Luminol

def get_luminol_anomalies(in_df):
    in_df['isAnomaly'] = 0
    detector = AnomalyDetector(in_df['y'].to_dict())
    anomalies = detector.get_anomalies()
    for anomaly in anomalies:
        for k in range(anomaly.get_time_window()[0], anomaly.get_time_window()[1]):
            in_df.at[k, 'isAnomaly'] = 1
    return in_df
```

#### Applying Anomaly Detection to Each l_ipn

```python
for i in range(df['l_ipn'].nunique()):
    t_df = get_luminol_anomalies(get_daily_aggregate_l_ipn(i))
    colors = {0: 'green', 1: 'red'}
    plt.figure(figsize=(10, 5))
    plt.plot(t_df['ds'], t_df['y'])
    plt.scatter(t_df['ds'], t_df['y'], c=t_df['isAnomaly'].apply(lambda x: colors[x]))
    plt.title(f'Forecast plot for l_ipn {i}')
    plt.show()
```
